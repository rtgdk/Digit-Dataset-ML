```python
import numpy as np
from sklearn.svm import SVC
rng = np.random.RandomState(1)
```

```python
X = rng.rand(100,10)
y = rng.binomial(1, 0.5, 100)
X_test = rng.rand(5, 10)
```

```python
from sklearn.svm import SVC
clf=SVC()
clf.set_params(kernel='linear').fit(X, y)
```
